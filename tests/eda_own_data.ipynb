```python
import pandas as pd

df = pd.read_parquet("../data/gold/tool_growth")
df2 = pd.read_parquet("../data/gold/activity_heatmap")
```

```python
# Vika DE verktyg har flest stars de senaste veckorna?
df = pd.read_parquet("../data/gold/tool_growth")
print(df.groupby("repo_name")["stars"].sum().sort_values(ascending=False).head(25))
```

```python
# NÄR är de communityt aktivt enlight mina params?

print(df2.sort_values("event_count", ascending=False).head(15))
```

```python
# NÄR är de communityt MINST aktiva enligt mina params?

print(df2.sort_values("event_count", ascending=True).head(15))
```
